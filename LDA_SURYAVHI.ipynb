## **IMPORTING REQUIRED LIBRARIES:-**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

## **READING THE DATASET:-**

```python
df=pd.read_csv('nba_final.csv')
```

```python
df.info()
```

```python
df
```

```python
## DELETING PLAYER NAME AND ID as we will deal with logistic regression.
df.drop(['Player.x','Player_ID'],axis=1,inplace=True)
```

```python
## checking for null values:-
df.isnull().sum()
```

```python
df['Pos2']=df['Pos2'].fillna('none')
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].median())
```

```python
df.fillna(0,inplace=True)
```

```python
#checking for null values again:-
df.isnull().sum().sum()
```

```python
# checking for duplicate values:-
df.duplicated().sum()
```

```python
df
```

## **FINAL DATASET BEFORE DOING OPERATION:-**

```python
df
```

## **LABEL ENCODING:-**

```python
obj_cols=df.select_dtypes('object').columns
obj_cols
```

```python
le=LabelEncoder()
```

```python
for col in obj_cols:
    df[col]=le.fit_transform(df[col])
```

```python
df
```

## **ML PROCESS:-**

```python
x=df.drop(columns='Play')
y=df['Play']
```

```python
y.value_counts()# checking the data is balanced or not
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

## **STANDERDIZATION/MEAN CENTERING:-**

```python
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

## **MODEL BUILDING:-**

```python
# Create an instance of the LogisticRegression class
model = LogisticRegression()
# Now you can fit the model to your data
model.fit(x_train_scaled, y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
#checking the accuracy:
roc_auc_score(y_test,y_pred)*100
```
