# **Importing Required Libraries:-**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

# **Data Reading:-**

```python
df=pd.read_csv('census-income_.csv')
df.head(10)
```

```python
df.info()
```

# **EDA Process:-**

```python
df.isnull().sum() # checking for duplicate values
```

```python
df.duplicated().sum()
```

```python
# Eliminating duplicated values:-
df.drop_duplicates(inplace=True)
```

# **Label Encoding:-**

```python
obj_columns=df.select_dtypes(include='object').columns
obj_columns
```

```python
df[' education'].value_counts()
```

# education Column is not that much important.

```python
df.drop(columns=' education',inplace=True)
```

```python
df['age'].value_counts()
```

# 'age' column is not that much important

```python
df.drop(columns='age',inplace=True)
```

```python
df.info()
```

```python
le=LabelEncoder()
for col in obj_columns:

  df[col]=le.fit_transform(df[col])
```

```python
sc=StandardScaler()
df[' fnlwgt']=sc.fit_transform(df[[' fnlwgt']])
df[' capital-gain']=sc.fit_transform(df[[' capital-gain']])
```

```python
df
```

# **ML PROCESSES:-**

```python
x=df.drop(columns='annual Income')
```

```python
y=df['annual Income']
```

```python
y.value_counts() #imbalance data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

## **LDA PROCESS:-**

```python
lda=LinearDiscriminantAnalysis()
lda.fit(x_train,y_train)
```

```python
lda.transform(x_test)
```

```python
lda.coef_
```

```python
strength=np.abs(lda.coef_)
```

```python
strength
```

```python
value=strength.flatten()
```

```python
value
```

```python
new_df=pd.DataFrame({'columns':x_train.columns,'strength':value})
```

```python
new_df
```

```python
best_col1=new_df.nlargest(5,'strength')['columns']
```

```python
best_col1
```

```python
best_col2=new_df.nlargest(8,'strength')['columns']
```

```python
best_col2
```

## **MODEL PREDECTION:-**

```python
x_train1=x_train.iloc[:,best_col1.index]
x_test1=x_test.iloc[:,best_col1.index]
```

```python
x_train2=x_train.iloc[:,best_col2.index]
x_test2=x_test.iloc[:,best_col2.index]
```

# Classification Models:-

```python
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=RandomForestClassifier()
```

# Logistic Regression:-

```python
model1.fit(x_train1,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model1.predict(x_test1)))
```

```python
model1.fit(x_train2,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model1.predict(x_test2)))
```

# Decision Tree Classifier:-

```python
model2.fit(x_train1,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model2.predict(x_test1)))
```

```python
model2.fit(x_train2,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model2.predict(x_test2)))
```

# Ramdom Forest Classifier:-

```python
model3.fit(x_train1,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model3.predict(x_test1)))
```

```python
model3.fit(x_train2,y_train)
print('Accuracy Score:',roc_auc_score(y_test,model3.predict(x_test2)))
```
